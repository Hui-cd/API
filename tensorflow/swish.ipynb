### tf.nn.swish(x) Swish Activation Function

$$\text{swish}(x) = x \cdot \sigma(x)$$
where $\sigma(x)$ is the sigmoid function defined as:

$$\sigma(x) = \frac{1}{1 + e^{-x}}$$
$Swis$是一个平滑的、非单调的函数，它已被发现可以帮助更深的模型，在ReLU可能失效的情况下表现更好

```python
import tensorflow as tf

# 假设 x 是一个 TensorFlow 张量
x = tf.constant([-3.0, -1.0, 0.0, 1.0, 3.0])

# 应用 Swish 激活函数
y = tf.nn.swish(x)

# 输出计算结果
print("Output:", y.numpy())
```

```python
import numpy as np
import matplotlib.pyplot as plt

def swish(x):
    return x * (1 / (1 + np.exp(-x)))

# 创建输入数据
x = np.linspace(-10, 10, 400)
y = swish(x)

# 绘制图像
plt.plot(x, y, label='swish(x) = x * sigmoid(x)')
plt.xlabel('x')
plt.ylabel('swish(x)')
plt.title('Swish Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```
